```python
!pip install flask flask-cors
from pyngrok import ngrok

# Tắt tất cả các tunnel đang chạy
ngrok.kill()

print("Đã tắt toàn bộ process ngrok cũ.")
from flask import Flask, request, jsonify
from underthesea import pos_tag, word_tokenize, classify
import json
from flask_cors import CORS

app = Flask(__name__)
CORS(app)

@app.route('/health', methods=['GET'])
def health():
    return jsonify({'status': 'ok'})

@app.route('/classify/pos', methods=['POST'])
def classify_pos():
    data = request.json
    text = data.get('text', '')
    result = pos_tag(text)
    return jsonify({
        'tokens': [word for word, tag in result],
        'pos_tags': [[word, tag] for word, tag in result],
        'success': True
    })

@app.route('/classify/tokenize', methods=['POST'])
def tokenize():
    data = request.json
    text = data.get('text', '')
    tokens = word_tokenize(text)
    return jsonify({
        'tokens': tokens,
        'success': True
    })

@app.route('/classify/text', methods=['POST'])
def classify_text():
    data = request.json
    text = data.get('text', '')
    result = classify(text)
    return jsonify({
        'classification': result,
        'success': True
    })

if __name__ == '__main__':
    # Expose URL công khai
    from flask import Flask
    from pyngrok import ngrok

    ngrok.set_auth_token("37kOuDP8SXUzcUwVz0Idz3EpOOu_6nA1e88jEujMWDNCA5qgv")  # Lấy từ https://dashboard.ngrok.com/
    public_url = ngrok.connect(5001)
    print(f"Flask server running at: {public_url}")

    app.run(host='0.0.0.0', port=5001, debug=False)
```
